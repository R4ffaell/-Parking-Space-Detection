```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load model yang sudah di-train
model = YOLO("runs/detect/train/weights/best.pt")  # ganti dengan path kamu

# Simpan prediksi dan label asli
y_true = []
y_pred = []

import os
from PIL import Image

# Ambil semua file gambar dari folder test
test_path = "/content/Parking-Space-1/test/images"
for file in os.listdir(test_path):
    if file.endswith(".jpg") or file.endswith(".png"):
        img_path = os.path.join(test_path, file)

        # Prediksi
        results = model(img_path)

        # Asumsi hanya 1 object per gambar (atau ambil object pertama aja)
        if results[0].boxes.cls.numel() > 0:
            pred_class = int(results[0].boxes.cls[0].item())
            y_pred.append(pred_class)

        # Ambil ground truth label (dari file label .txt)
        label_file = img_path.replace("/images/", "/labels/").replace(".jpg", ".txt").replace(".png", ".txt")
        if os.path.exists(label_file):
            with open(label_file, "r") as f:
                line = f.readline()
                if line:
                    true_class = int(line.split(" ")[0])
                    y_true.append(true_class)
```

```python
y_true = []
y_pred = []

test_path = "/content/Parking-Space-1/test/images"

for file in os.listdir(test_path):
    if file.endswith(".jpg") or file.endswith(".png"):
        img_path = os.path.join(test_path, file)

        # Path label
        label_path = img_path.replace("/images/", "/labels/").replace(".jpg", ".txt").replace(".png", ".txt")

        # Cek apakah label file-nya ada
        if os.path.exists(label_path):
            # Baca ground truth
            with open(label_path, "r") as f:
                lines = f.readlines()
                if len(lines) == 0:
                    continue  # skip jika tidak ada label

                true_class = int(lines[0].split(" ")[0])  # hanya ambil objek pertama
                y_true.append(true_class)

            # Prediksi
            results = model(img_path)
            boxes = results[0].boxes

            if boxes.cls.numel() > 0:
                pred_class = int(boxes.cls[0].item())  # ambil prediksi object pertama
                y_pred.append(pred_class)
            else:
                y_pred.append(-1)  # jika model tidak mendeteksi, masukkan label -1
                print(f"No prediction for {file}, assigning -1.")
```
